```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
%matplotlib inline
```

```python
df = pd.DataFrame({
    'Tiempo': [0,1,2,3,4,5],
    'Temperatura': [80,70,62,55,49,44]
})

df.head()
```

```python
df.describe()
```

```python
plt.figure()
plt.plot(df['Tiempo'], df['Temperatura'])
plt.xlabel('Tiempo')
plt.ylabel('Temperatura')
plt.title('Ejemplo de enfriamiento')
plt.show()
```
